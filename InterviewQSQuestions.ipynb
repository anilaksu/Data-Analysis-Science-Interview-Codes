```python
# Object oriented programming perks:
# 1) Modularity: Object-oriented programming provides object-based separation of code and behavior
# 2) Extensibility: Objects can be extended to include new behaviours
# 3) Reusability: objects can be reused everywhere in the application

# this is the class - so the blueprint
# Encapsulation - it is a fundamental feature of OOP
# we are able to group certain variables and functions
class Person:
  name = 'Kevin'
  # so this is the so-called constructor (initialization block)
  def __init__(self, name, age = 20):
    self.name = name
    self.age = age

  def show_name(self):
    print('My name is %s and age is %s ' % (self.name, self.age))

# we can instantiate a class: and p is the object
p = Person(name = 'Mahmut', age = 36)
p.show_name()
```

```python
# Class variables: shared by all object instances of the given class - the class own the variable
# Instance variable: the value of these variables can be different depending on the instance
class Person:
  gender = 'Male'  # This is a class variable which is the same for every instantiation
  person_list = [] # It is a dynamic class variable
  # so this is the so-called constructor (initialization block)
  def __init__(self, name, age = 20):
    self.name = name
    self.age = age

  def show_name(self):
    print('My name is %s and age is %s and gender is %s ' % (self.name, self.age, self.gender))

  def add_value(self, value):
    self.person_list.append(value)

p1 = Person(name = 'Kevin')
p2 = Person(name = 'Joe', age = 15)

p1.add_value('person1')
p1.add_value('person2')

p1.show_name()
p2.show_name()

print(p1.person_list)
print(p2.person_list)
```

```python
# Inheritance allows us to define a class (child class or derived class) that inherits all the properties and functions from another class
class Algorithm:

  # class variable
  general_message = 'This has something to do with algorithms...'

class SortingAlgorithm(Algorithm):

  # class variable
  sorting_message = 'This is a sorting algorithm...'

sorting_algorithm = SortingAlgorithm()
print(sorting_algorithm.general_message)
```

```python
# Super keyword
class User:

  # so this is the so-called constructor (initialization block)
  def __init__(self, name, age = 20):
    self.name = name
    self.age = age

  def show_user(self):
    print("The user is: %s and the age is: %s" % (self.name, self.age))

  def show_info(self):   # This is from the parent class
    print("This is from the parent...")

class FacebookUser(User):

  # super keywork is going to call the parent class
  def __init__(self, name, age):
      super().__init__(name, age)

  def show_info(self):   # This is from the child class (Function Override: it is present in the parent class)
    print("This is from the child...")

facebook_user = FacebookUser('Adam Smith', 35)
facebook_user.show_user()
facebook_user.show_info()
```

```python
# Polymorphism is the ability of an object to take on many forms
# In object-oriented programming, it can be achieved by inheritance
class SortingAlgorithm: # Parent Class

  def sort(self):
    pass


class InsertionSort(SortingAlgorithm):

  # function override
  def sort(self):
    print('Insertion sort is running...')

class SelectionSort(SortingAlgorithm):

  # function override
  def sort(self):
    print('Selection sort is running...')

class QuickSort(SortingAlgorithm):

  # function override
  def sort(self):
    print('Quick sort is running...')

# We ask the user what sorting algortihm to use
num = input("Please, select a sortin algorithm")
num = int(num)

# this is how we create a NULL object
sorting_algorithm = None

if num == 0:
  sorting_algorithm = InsertionSort()
elif num == 1:
  sorting_algorithm = SelectionSort()
elif num == 2:
  sorting_algorithm = QuickSort()
else:
  num = input("Please enter valid number")
  num = int(num)

# this object can take on many forms
sorting_algorithm.sort()
```

```python
# Vehicle Parent Class and Bus and Car Child Class
# Super keyword
class Vehicle:

  # so this is the so-called constructor (initialization block)
  def __init__(self, brand, year = 2020):
    self.brand = brand
    self.year = year

  def show_vehicle(self):
    print("Vehicle is: %s and the year is: %s" % (self.brand, self.year))

  def fuel_consumption(self):
    print("Vehicle is: %s and the year is: %s" % (self.brand, self.year))

class Bus(Vehicle):

  # super keywork is going to call the parent class
  def __init__(self, brand, year):
      super().__init__(brand, year)

  def fuel_consumption(self):
    print("This bus burns 2 gallons per mile")

class Car(Vehicle):

  # super keywork is going to call the parent class
  def __init__(self, brand, year):
      super().__init__(brand, year)

  def fuel_consumption(self):
    print("This bus burns 1 gallon per mile")

car1 = Car(brand = 'Ford', year = 2021)
car1.show_vehicle()
car1.fuel_consumption()
```

```python
# Comparing objects
class Person:

  def __init__(self, name, age):
    self.name = name
    self.age = age
  def __eq__(self, other): # We override equal function so that we can define the equivalance
    return self.age == other.age

  def __lt__(self, other): # We override equal function so that we can define less than
    return self.age < other.age

# how to compare the objects?
p1 = Person('Adam', 32)
p2 = Person('Joe', 34)

print(p1 == p2)
print(p1 < p2)
```

```python
# Here we convert paranthesis into its end result and simplify the expersion
import operator
ops = { "+": operator.add, "-": operator.sub }

def direct_sum(input):
  result = 0    # here we initiate the end result
  subresult = 0 # here we store intermediate results
  op = '+'      # Default operation for the first entry
  for i in range(0,len(input)):
    if input[i] in ['+','-']:  # Here we check the first entry if it a number or operation
      op = input[i]
    elif input[i] not in ['+','-'] and i == 0:      # Here we specify the operation for the first entry
      subresult = int(input[i])
      result = result + subresult
    elif input[i] not in ['+','-']:
      subresult = int(input[i])
      result = ops[op](result,subresult)
  return result

def shrink_paranthesis(input):  # Here we shrink paranthesis into values

  try:
     p_s = input.index('(')      # paranthesis start indice
     p_e = input.index(')')      # paranthesis start indice
     return  direct_sum(input[0:p_s] + str(direct_sum(input[p_s+1:p_e])) + input[p_e+1:])  # Here we sum the converted string
  except:
     p_s = 0                     # default paranthesis start indice
     p_e = len(input)-1          # default paranthesis start indice
     return  str(direct_sum(input))  # Here we sum the string as is


```

## 6.3.1 Holt-Winters' seasonal method

\begin{align*}
  \hat{y}_{t+h|t} &= \ell_{t} + hb_{t} + s_{t+h-m(k+1)} \\
  \ell_{t} &= \alpha(y_{t} - s_{t-m}) + (1 - \alpha)(\ell_{t-1} + b_{t-1})\\
  b_{t} &= \beta^*(\ell_{t} - \ell_{t-1}) + (1 - \beta^*)b_{t-1}\\
  s_{t} &= \gamma (y_{t}-\ell_{t-1}-b_{t-1}) + (1-\gamma)s_{t-m},
\end{align*}

```python
# Here we fit the model get the summary
model = HWES(np.asarray(df_train), seasonal_periods=365, trend='add', seasonal='mul')
fitted = model.fit()
print(fitted.summary())
```

## 6.5 Simulating a roulette strategy
In betting, there is a strategy known as the Martingale betting system where the gambler simply doubles their bet after each loss, so their first eventual win covers all the previous losses + a profit equal to the original stake. This strategy is designed for games with roughly 50/50 odds, such as roulette (where the odds are slightly less than 50/50). For example, if we lose our first roulette \$1 bet, we will bet \$2 on the next bet trying to make up for the previous loss. If we lose that bet, we will then bet \$4 and so on until we finally win a roll and reset to our initial bet of \$1 (taking home the \$1 profit of the original stake).
    
Given this information, write a function that can simulate the bankroll of an individual who employs this strategy playing roulette over time. You can assume the gambler starts with $5000 and makes an initial bet of \$1.

```python
import numpy as np
# Here we define the Martingale betting system
class MartingalBetting:

  # so this is the so-called constructor (initialization block)
  def __init__(self, budget, initial_bet):
    self.budget = budget
    self.initial_bet = initial_bet
    self.current_bet = initial_bet           # Here we simply define our current bet to be updated during the game
  # here we generate corresponding letters to each number on a telephone pad
  def play_round(self):
      if(np.random.choice(2, 1, replace=False, p=[0.5, 0.5]) == 1): # Here we assign random integer value with weights
        self.budget = self.budget + self.current_bet                # Here we adjust our budget according to the result of our bet
        self.current_bet = 0                                        # Here we set current bet to zero as we finish the game
      else:
        self.budget = self.budget - self.current_bet                # Here we adjust our budget according to the result of our bet
        self.current_bet = 2 * self.current_bet                     # Here we set currrent bet
```

```python
if __name__ == '__main__':

  budget = 5000                                       # Our initial budget
  initial_bet = 1                                     # Our initial bet
  budget_array = []                                   # It is a dynamic class variable to store our budget information
  bet = []                                            # It is a dynamic class variable to store our bet information

  our_betting = MartingalBetting(budget,initial_bet)  # Here we initialize our betting system
  budget_array.append(our_betting.budget)             # Here we add the initial budget to list
  bet.append(our_betting.current_bet)                 # Here we add the current budget to list

  while our_betting.budget < budget+initial_bet and our_betting.budget > 0: # Here we set the end conditions which are either we win our game as planned or we left with no budget
    our_betting.play_round()                          # Here we play a round
    budget_array.append(our_betting.budget)           # Here we add the current budget to list
    bet.append(our_betting.current_bet)               # Here we add the current budget to list


  print("Our budget over the rounds: ", budget_array)
  print("Our bet over the rounds: ",  bet)
```

Suppose you're working for an accounting firm and looking to automate some tax estimates for clients. Under the relevant tax laws, an individual is taxed at 45% for any annual earnings >= 2M USD. For any earnings < 2M USD, the individual is taxed at a lower rate of 35%.

```python
# Here we define post-tax earnings
def getPostTaxEarnings(pretax_earnings, lim):
  conditions = [
        (pretax_earnings <= lim),
        (pretax_earnings > lim)]

  #define results
  results = [pretax_earnings * (1-0.35) ,   # Post tax earning if pretax earning below the limit
                  (pretax_earnings -lim) * (1-0.45) + lim * (1-0.35)] # Post tax earning if pretax earning below the limit
  return np.select(conditions, results)
```

```python
# Here we check if the string A is a subsequence of string B
import numpy as np

def check_subsequence(string_A,string_B):

  string_A = list(string_A)           # String A as list
  string_B = list(string_B)           # String B as list
  index = []                          # Index of String A elements in String B

  for i in range(0,len(string_A)):
    index.append(string_B.index(string_A[i])) # Index of string A elements

  #!!! Note that we specify they are not the same array with [:], if not every operation we do will also affect index_sorted list as well
  index_unsorted = index[:]          # Here we keep unsort array
  index.sort()                       # Here we sort the index arrayso that we can check if it changes when we sort it

  if index_unsorted == index:
    return  True                               # The order is the same so it is a subsequence
  else:
    return False                               # The order is not the same so it is not a subsequence
```

**Input Format**

The input consists of three lines. The first line contains the integer $N$, denoting the length of the list. The next line consists of $K$ space-separated lowercase English letters, denoting the elements of the list.

The third and the last line of input contains the integer $K$, denoting the number of indices to be selected.

**Output Format**

Output a single line consisting of the probability that at least one of the  indices selected contains the letter:'$a$'.

**Note:** The answer must be correct up to 3 decimal places.

**Constraints**
\begin{equation}

1 \leq K \leq N
\end{equation}

All the letters in the list are lowercase English letters.

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import sys
import re
from itertools import combinations

lines = [] # Here we initiate lines as list so that we will read all at once
for line in sys.stdin:
    lines.append(line.rstrip())

N = int(lines[0])                            # Number of letters
Letters = list(re.findall(r"\w", lines[1]))  # Letters
K = int(lines[2])                            # Index

comb = combinations(Letters, 2)        # All combinations
comb_count = 0                         # Number of all combinations
letter_count = 0                       # Count of combinations with indexed letter

for i in list(comb):
    comb_count +=1
    if Letters[K-1] in list(i):
        letter_count +=1

print("{:.3f}".format(round(letter_count/comb_count, 3)))
```

### 6.18 Polynomial Regression

The Problem

Charlie wants to purchase office-space. He does a detailed survey of the offices and corporate complexes in the area, and tries to quantify a lot of factors, such as the distance of the offices from residential and other commercial areas, schools and workplaces; the reputation of the construction companies and builders involved in constructing the apartments; the distance of the offices from highways, freeways and important roads; the facilities around the office space and so on.

Each of these factors are quantified, normalized and mapped to values on a scale of 0 to 1. Charlie then makes a table. Each row in the table corresponds to Charlie's observations for a particular house. If Charlie has observed and noted F features, the row contains F values separated by a single space, followed by the office-space price in dollars/square-foot. If Charlie makes observations for H houses, his observation table has (F+1) columns and H rows, and a total of (F+1) * H entries.

Charlie does several such surveys and provides you with the tabulated data. At the end of these tables are some rows which have just F columns (the price per square foot is missing). Your task is to predict these prices. F can be any integer number between 1 and 5, both inclusive.

There is one important observation which Charlie has made.

    The prices per square foot, are (approximately) a polynomial function of the features in the observation table. This polynomial always has an order less than 4

Input Format

The first line contains two space separated integers, F and N. Over here, F is the number of observed features. N is the number of rows for which features as well as price per square-foot have been noted.
This is followed by a table having F+1 columns and N rows with each row in a new line and each column separated by a single space. The last column is the price per square foot.

The table is immediately followed by integer T followed by T rows containing F columns.

Constraints

1 <= F <= 5
5 <= N <= 100
1 <= T <= 100
0 <= Price Per Square Foot <= 10^6 0 <= Factor Values <= 1

Output Format

T lines. Each line 'i' contains the predicted price for the 'i'th test case.

```python
def Extract(lst):
    return [item[:-1] for item in lst], [item[-1] for item in lst]
```

```python
# Relevant libraries
import numpy
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

data = []; # Price and feature data

numbers = numpy.array([int(elem) for elem in input().split()]); # Numbers

F = numbers[0]; # Features
N = numbers[1]; # Number of entries

for i in range(N):
  data.append([float(elem) for elem in input().split()]); # Features and prices

# Here we separate data
X, y = Extract(data);
#generate a model of polynomial features
poly = PolynomialFeatures(degree=3)

# Transform the x data for proper fitting (for single variable type it returns,[1,x,x**2])
X_ = poly.fit_transform(X)

# Generate the regression object
clf = linear_model.LinearRegression()
# Perform the actual regression
clf.fit(X_, y)

# Input for the test cases
numbers = int(input()); # Number of test cases

X_test = [];

# Test features
for i in range(numbers):
  X_test.append([float(elem) for elem in input().split()]); # Test features

# Transform the test data
X_test_ = poly.fit_transform(X_test)

# Prediction
y_preds = clf.predict(X_test_ );

for y_pred in y_preds:
    print("{0:0.2f}".format(y_pred))
```
